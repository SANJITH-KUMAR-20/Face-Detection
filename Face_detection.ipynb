```python
import numpy as np
import cv2 as cv
from tensorflow.keras.models import load_model
import tensorflow as tf
```

```python
my_model = load_model("himodel.h5",compile = False)
```

```python
cap = cv.VideoCapture(0)
while cap.isOpened():
    _ , frame = cap.read()
    frame = frame[50:500,50:500,:]
    framey = cv.cvtColor(frame,cv.COLOR_BGR2RGB)
    resized = tf.image.resize(framey,(120,120))
    predicted = my_model.predict(np.expand_dims(resized/255.0,0))
    sample_coords = predicted[1][0]

    if predicted[0]>0.5:
        cv.rectangle(frame, 
                      tuple(np.multiply(sample_coords[:2], [450,450]).astype(int)),
                      tuple(np.multiply(sample_coords[2:], [450,450]).astype(int)), 
                            (255,0,0), 2)
        cv.rectangle(frame, 
                      tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int), 
                                    [0,-30])),
                      tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int),
                                    [80,0])), 
                            (255,0,0), -1)
        cv.putText(frame, 'FACE', tuple(np.add(np.multiply(sample_coords[:2], [450,450]).astype(int),
                                               [0,-5])),
                    cv.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv.LINE_AA)
        cv.imshow('EyeTrack', frame)
    
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv.destroyAllWindows()
    
```
